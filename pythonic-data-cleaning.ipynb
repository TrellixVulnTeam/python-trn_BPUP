```python
caracteres_indesejados = ['[', ',', '-', ']', '.', '?', '/']

def limpa_datas(item):
    data_publicacao = str(item.loc['data_publicacao'])
    
    if data_publicacao == 'nan' or data_publicacao[0] == '[':
        return np.NaN
    
    for caractere in caracteres_indesejados:
        if caractere in data_publicacao:
            indice_caractere = data_publicacao.find(caractere)
            data_publicacao = data_publicacao[:indice_caractere]
    
    return data_publicacao

dataframe_images_book['data_publicacao'] = dataframe_images_book.apply(limpa_datas, axis = 1)
```
